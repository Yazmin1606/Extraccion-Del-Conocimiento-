```python
# Importamos las librerías necesarias
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
# Lista de influencers relacionados con turismo y estilo de vida en Cancún
influencers = [
    "travel_cancun",
    "beach_life",
    "cancun_vibes",
    "mexico_travel",
    "sunset_beach"
]
```

```python
# Lista de influencers relacionados con turismo y estilo de vida en Cancún
tipos_post = ["foto", "reel", "historia", "video"]
```

```python
# Lista donde se almacenarán los registros del dataset
datos = []
```

```python
# Lista donde se almacenarán los registros del dataset
for i in range(10000):
    datos.append({
        "influencer": random.choice(influencers),
        "likes": random.randint(100, 50000),
        "comentarios": random.randint(5, 5000),
        "fecha": "2025-05-01",
        "tipo_post": random.choice(tipos_post)
    })
```

```python
# Conversión de los datos a un DataFrame de Pandas
df = pd.DataFrame(datos)
```

```python
df.info()
```

```python
# Identificar la publicación con mayor cantidad de likes
df["likes"].mean()
```

```python
# Identificar la publicación con mayor cantidad de comentarios
df["comentarios"].mean()
```

```python
# Mostrar cuántas publicaciones existen de cada tipo (foto, reel, historia y video)
df["tipo_post"].value_counts().plot(kind="bar")
```

```python
# Contar cuántas publicaciones hay de cada tipo
df["tipo_post"].value_counts()
```

```python
# Graficar la distribución de tipos de publicaciones
df["tipo_post"].value_counts().plot(kind="bar")
```
